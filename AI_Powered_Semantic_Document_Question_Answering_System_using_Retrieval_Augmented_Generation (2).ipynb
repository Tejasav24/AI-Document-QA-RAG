#AI-Powered Semantic Document Question Answering System using Retrieval-Augmented Generation

##Problem Statement

In the modern digital era, a large amount of textual data is generated and stored in documents, reports, and online resources. Extracting relevant information from these large text collections is often difficult and time-consuming. Traditional keyword-based search systems fail to capture the semantic meaning of user queries, which often leads to inaccurate or irrelevant results.

The objective of this project is to develop an AI-powered document question answering system that can understand the semantic meaning of user queries and retrieve the most relevant information from documents. The system uses sentence embeddings and vector similarity search to identify relevant document segments and then utilizes a transformer-based language model to generate accurate answers. This approach improves information retrieval efficiency and provides context-aware responses to user queries.

```python
import faiss
import numpy as np

embeddings = np.array(embeddings)

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(embeddings)

print("Total vectors:", index.ntotal)
```

##Semantic Query Processing and Document Retrieval

```python
def ask_question(query):

    query_embedding = model.encode([query])

    distances, indices = index.search(np.array(query_embedding), k=3)

    results = []

    for i in indices[0]:
        results.append(chunks[int(i)])

    return "\n\n".join(results)
```

##End-to-End AI Document Chat System Implementation

```python
import PyPDF2
import faiss
import numpy as np
import gradio as gr

from sentence_transformers import SentenceTransformer
from google.colab import files

print("Upload your PDF files")

uploaded = files.upload()

print("Reading PDFs...")

text = ""

for file in uploaded.keys():

    reader = PyPDF2.PdfReader(file)

    for page in reader.pages:
        page_text = page.extract_text()

        if page_text:
            text += page_text


print("Creating text chunks...")

chunk_size = 500

chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

print("Total chunks:", len(chunks))


print("Loading embedding model...")

model = SentenceTransformer("all-MiniLM-L6-v2")


print("Creating embeddings...")

embeddings = model.encode(chunks)

embeddings = np.array(embeddings)


print("Building vector database...")

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(embeddings)

print("Vector database ready")


def search_docs(query):

    query_embedding = model.encode([query])

    distances, indices = index.search(np.array(query_embedding), k=3)

    results = []

    for i in indices[0]:
        results.append(chunks[int(i)])

    return "\n\n".join(results)


def chat(query, history):

    response = search_docs(query)

    history.append((query, response))

    return history, history


print("Launching DocuMind AI...")

with gr.Blocks() as demo:

    gr.Markdown("# DocuMind AI")
    gr.Markdown("Chat with your PDF documents")

    chatbot = gr.Chatbot()

    msg = gr.Textbox(label="Ask a question")

    state = gr.State([])

    msg.submit(chat, [msg, state], [chatbot, state])


demo.launch()
```

##Smart Text Chunking with Overlap

```python
chunk_size = 500
overlap = 100

chunks = []

for i in range(0, len(text), chunk_size - overlap):
    chunk = text[i:i+chunk_size]
    chunks.append(chunk)

print("Total chunks:", len(chunks))
```

##Top-K Semantic Retrieval Optimization

```python
# ---------------------------
# Function: search top-k relevant chunks
# ---------------------------
def search_docs(query, top_k=5):
    """
    Parameters:
        query  : str  -> user input question
        top_k  : int  -> number of top results to return

    Returns:
        str -> concatenated text of top-k similar chunks
    """
    global index, chunks, model

    if not chunks or index is None:
        return "PDF not processed yet. Please upload and process PDFs first."

    # 1. Create embedding for query
    query_embedding = model.encode([query])

    # 2. Search in FAISS index
    distances, indices = index.search(np.array(query_embedding), k=top_k)

    # 3. Retrieve corresponding text chunks
    results = [chunks[int(i)] for i in indices[0]]

    return "\n\n".join(results)
```

##Text-to-Text QA Pipeline Using Flan-T5

```python
# ---------------------------
# Heading: Flan-T5 Text-to-Text QA Pipeline
# ---------------------------

from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, pipeline

# Load model & tokenizer
model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Pipeline for text2text generation (works in Transformers now)
qa_pipeline = pipeline(
    task="text-generation",   # fixed task
    model=model,
    tokenizer=tokenizer
)

# Function to generate answer
def generate_answer(query, context=None, max_length=200):
    """
    Generates answer for a query.
    Optionally, context from PDF can be prepended to question for better results.
    """
    if context:
        prompt = f"Context: {context}\n\nQuestion: {query}\nAnswer:"
    else:
        prompt = f"Question: {query}\nAnswer:"

    result = qa_pipeline(prompt, max_length=max_length, do_sample=True)
    return result[0]['generated_text']

# Example
query = "Explain machine learning in simple words"
context = "Machine learning is a subset of AI that allows computers to learn patterns from data..."
answer = generate_answer(query, context)
print(answer)
```

##DocuMind AI - PDF Semantic Search + Flan-T5 QA


```python
# ---------------------------
# Heading: DocuMind AI - PDF Semantic Search + Flan-T5 QA
# ---------------------------

# Install required packages (Colab)
!pip install sentence-transformers faiss-cpu gradio PyPDF2 transformers --quiet

# Imports
import PyPDF2
import faiss
import numpy as np
import gradio as gr
from sentence_transformers import SentenceTransformer
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
from google.colab import files

# ---------------------------
# Upload PDFs
# ---------------------------
print("Upload your PDF files")
uploaded = files.upload()

all_text = ""
for file in uploaded.keys():
    reader = PyPDF2.PdfReader(file)
    for page in reader.pages:
        text = page.extract_text()
        if text:
            all_text += text

# ---------------------------
# Chunk the text
# ---------------------------
chunk_size = 500
chunks = [all_text[i:i+chunk_size] for i in range(0, len(all_text), chunk_size)]
print("Total chunks:", len(chunks))

# ---------------------------
# Create embeddings (Semantic Search)
# ---------------------------
model_embed = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model_embed.encode(chunks)
embeddings = np.array(embeddings)
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)
print("Vector DB ready")

# ---------------------------
# Flan-T5 QA Model
# ---------------------------
flan_model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(flan_model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(flan_model_name)

def generate_answer(query, context=None, max_length=200):
    """Use Flan-T5 to generate answer given query and context"""
    if context:
        prompt = f"Context: {context}\n\nQuestion: {query}\nAnswer:"
    else:
        prompt = f"Question: {query}\nAnswer:"

    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=max_length)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# ---------------------------
# Top-K Semantic Retrieval
# ---------------------------
def search_docs(query, top_k=3):
    """Retrieve top-K relevant chunks from PDF using FAISS"""
    q_embed = model_embed.encode([query])
    distances, indices = index.search(np.array(q_embed), k=top_k)
    results = [chunks[i] for i in indices[0]]
    return "\n\n".join(results)

# ---------------------------
# Combined QA Function
# ---------------------------
def docuqa(query):
    context = search_docs(query, top_k=3)
    answer = generate_answer(query, context)
    return answer

# ---------------------------
# Gradio Interface
# ---------------------------
with gr.Blocks() as demo:
    gr.Markdown("# DocuMind AI")
    gr.Markdown("Ask questions from your uploaded PDFs")
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Ask a question")
    state = gr.State([])
    def chat_fn(query, history):
        answer = docuqa(query)
        history.append((query, answer))
        return history, history
    msg.submit(chat_fn, [msg, state], [chatbot, state])

demo.launch()
```

##Retrieval-Augmented Answer Generation using Flan-T5

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Model & tokenizer
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")

def generate_answer(query, context):
    # Context + question prompt
    prompt = f"Answer the question based on the context below:\n{context}\nQuestion: {query}"

    # Tokenize input
    inputs = tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True)

    # Generate answer
    outputs = model.generate(**inputs, max_length=150)

    # Decode output
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return answer
```

##End-to-End Retrieval Augmented Question Answering Pipeline

```python
def ask_question(query):

    # Step 1: Retrieve relevant chunks
    context = search_docs(query, top_k=5)

    # Step 2: Generate answer using LLM
    answer = generate_answer(query, context)

    print("Question:", query)
    print("\nRetrieved Context:\n", context)
    print("\nFinal Answer:\n", answer)

    return answer
```

##Project Summary

This project develops an intelligent document-based question answering system using modern Natural Language Processing techniques. The system processes large textual data by dividing it into smaller chunks, generating semantic embeddings using transformer-based sentence encoders, and storing these embeddings in a vector database for efficient similarity search.

When a user submits a query, the system converts the query into an embedding and retrieves the most semantically relevant document chunks using vector similarity search. These retrieved contexts are then passed to a transformer-based language model to generate a concise and context-aware answer.

The project demonstrates how retrieval-augmented generation can enhance information access by combining semantic search with language generation. This approach improves accuracy, scalability, and user experience compared to traditional keyword-based retrieval systems
